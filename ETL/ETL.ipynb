# **`Transformation`**

```python
callCenter['Hour'] = callCenter['Date'].dt.hour

callCenter.head()
```

```python
callCenter['Date'] = pd.to_datetime(callCenter['Date'].dt.strftime("%Y-%m-%d"), format="%Y-%m-%d")

callCenter.head()
```

```python
callCenter['AvgTalkDuration'] = callCenter['AvgTalkDuration'].apply(lambda n : str(n)[:8])

callCenter['AvgTalkDuration'] = pd.to_timedelta(callCenter['AvgTalkDuration']).dt.total_seconds().astype(int)

callCenter.rename(columns={"AvgTalkDuration":"AvgTalkDuration (seconds)"}, inplace=True)

callCenter.head()
```

```python
callCenter[['Agent', 'Department', 'Answered (Y/N)', 'Resolved']] = callCenter[['Agent', 'Department', 'Answered (Y/N)', 'Resolved']].map(lambda n : n.strip())

callCenter.head()
```

```python
callCenter.rename(
    columns={"Answered (Y/N)":"Answered",
             "Speed of Answer":"SpeedOfAnswer",
             "AvgTalkDuration (seconds)":"AvgTalkDuration(Seconds)",
             "Satisfaction rating":"SatisfactionRating",
             "Call Id":"CallId"},
    inplace=True
)

callCenter.head()
```
